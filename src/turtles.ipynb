```python
t = Template("""
APPMODE stdAppmode {};

{% for task in tasks %}
  ALARM {{ task.name }}Alarm {
    COUNTER= SystemCounter;
    ACTION = ACTIVATETASK {
      TASK = {{ task.name }};
    };
    AUTOSTART = TRUE {
      ALARMTIME = {{ task.alarmtime }};
      CYCLETIME = {{ task.cycletime }};
      APPMODE = stdAppmode;
    };
  };
  TASK {{ task.name }} {
    PRIORITY = {{ task.alarmtime }};
    AUTOSTART = FALSE;
    ACTIVATION = 1;
    SCHEDULE = FULL;
    STACKSIZE = {{ task.stacksize }};
  };
{% endfor %}
""")
```

```python
from copy import deepcopy
```

```python
tasks = list()
task_default = {
    "alarmtime": 5000,
    "cycletime": 10,
    "priority": 20,
    "stacksize": 256,
}

cycletime=10
for pin in range(8, 14):
    task = deepcopy(task_default)
    task["name"]=f"blinkTaskPin{pin}"
    task["cycletime"]=cycletime
    tasks.append(task)
    cycletime=cycletime*2
print(t.render(tasks=tasks))
```
